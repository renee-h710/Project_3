```python
import pandas as pd 
import sqlite3
```

```python
#Call Census file 
census_file = 'Resources/Census_data.csv'
census_file = pd.read_csv(census_file)
census_file.head()
```

```python
#Drop the "Unnamed: 0" column
census_file.drop(["Unnamed: 0"], axis=1, inplace=True) 
census_file.head()

#Drop Margin of Error and Percent Margin of Error columns  
margin = census_file.loc[:,census_file.columns.str.contains("Margin of Error")]
census_file.drop(margin, axis=1, inplace=True) 

#Drop label(Grouping)
label_grouping = census_file.loc[:,census_file.columns.str.contains("Label")]
census_file.drop(label_grouping, axis=1, inplace=True)

#Replace !! with _ 
census_file.columns = census_file.columns.str.replace('!!','_')
census_file.head()
```

```python
#Pull out estimates 
estimate = census_file.loc[:,census_file.columns.str.contains("Estimate")]
estimate.head()
```

```python
#Pull out percentages
percentage = census_file.loc[:,census_file.columns.str.contains("Percent")]
percentage.head()
```

```python
#Occupancy percentage df 
occupied_p = percentage.iloc[1]
vacant_p = percentage.iloc[2]

occupancy_per_df = pd.DataFrame({"Occupied": occupied_p, 
        "Vacant": vacant_p
        })

occupancy_per_df["Occupied"] = occupancy_per_df["Occupied"].str.replace("%","").astype(float)
occupancy_per_df["Vacant"] = occupancy_per_df["Vacant"].str.replace("%","").astype(float)

occupancy_per_df.head()
```

```python
#Value estimate df
less_than_fifty = estimate.iloc[80]
fifty = estimate.iloc[81]
hundred = estimate.iloc[82]
hundredfifty = estimate.iloc[83]
twohundred = estimate.iloc[84]
threehundred = estimate.iloc[85]
fivehundred = estimate.iloc[86]
million = estimate.iloc[87]

value_est_df = pd.DataFrame({
        "Less than $50,000": less_than_fifty, 
        "$50,000 to $99,999": fifty, 
        "$100,000 to $149,999": hundred, 
        "$150,000 to $200,999": hundredfifty, 
        "$300,000 to $499,999": threehundred, 
        "$500,000 to $999,999": fivehundred, 
        "$1,000,000 or more": million
        })

value_est_df["Less than $50,000"] = value_est_df["Less than $50,000"].str.replace(",","").astype(int)
value_est_df["$50,000 to $99,999"] = value_est_df["$50,000 to $99,999"].str.replace(",","").astype(int)
value_est_df["$100,000 to $149,999"] = value_est_df["$100,000 to $149,999"].str.replace(",","").astype(int)
value_est_df["$150,000 to $200,999"] = value_est_df["$150,000 to $299,999"].str.replace(",","").astype(int)
value_est_df["$300,000 to $499,999"] = value_est_df["$300,000 to $499,999"].str.replace(",","").astype(int)
value_est_df["$500,000 to $999,999"] = value_est_df["$500,000 to $999,999"].str.replace(",","").astype(int)
value_est_df["$1,000,000 or more"] = value_est_df["$1,000,000 or more"].str.replace(",","").astype(int)

value_est_df.head()
```
