```python
import pandas as pd 
```

```python
#Call Census file 
census_file = 'Resources/Census_data.csv'
census_file = pd.read_csv(census_file)
census_file.head()
```

```python
#Drop the "Unnamed: 0" column
census_file.drop(["Unnamed: 0"], axis=1, inplace=True) 
census_file.head()

#Drop Margin of Error and Percent Margin of Error columns  
margin = census_file.loc[:,census_file.columns.str.contains("Margin of Error")]
census_file.drop(margin, axis=1, inplace=True) 

#Drop label(Grouping)
label_grouping = census_file.loc[:,census_file.columns.str.contains("Label")]
census_file.drop(label_grouping, axis=1, inplace=True)

#Replace !! with _ 
census_file.columns = census_file.columns.str.replace('!!','_')
census_file.head()
```

```python
#Pull out estimates 
estimate = census_file.loc[:,census_file.columns.str.contains("Estimate")]
estimate.head()
```

```python
#Pull out percentages
percentage = census_file.loc[:,census_file.columns.str.contains("Percent")]
percentage.head()
```

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///Resources/data.sqlite',echo=False)
# Enter all tables into the DB
# import sqlite3

```

```python
# create occupancy table
engine.execute('DROP TABLE "Occupancy_percentage";')
engine.execute('CREATE TABLE "Occupancy_percentage" ('
               'State VARCHAR NOT NULL,'
               'Occupied FLOAT, '
               'Vacant FLOAT, '
               'PRIMARY KEY (State));')


# create Value table
engine.execute('DROP TABLE "Value_estimate";')
engine.execute('CREATE TABLE "Value_estimate" ('
               'state VARCHAR NOT NULL,'
               'lessThan50 INTEGER, '
               'fifty INTEGER, '
               'hundred INTEGER, '
               'hundredFifty INTEGER, '
               'threeHundred INTEGER, '
                  'fiveHundred INTEGER, '
                'million INTEGER, '
               'PRIMARY KEY (state));')

# create Gross Rent table
engine.execute('DROP TABLE "Gross_rent_percentage";')
engine.execute('CREATE TABLE "Gross_rent_percentage" ('
      'state VARCHAR NOT NULL,'
      'lessThan15 FLOAT,'
      'fifteen FLOAT,'
      'twenty FLOAT,'
      'twentyfive FLOAT,'
      'thirty FLOAT,'
      'thirtyfive FLOAT,'
      'PRIMARY KEY (state));')
```

```python
#Connect Used DataFrames to SQLite DB
gross_rent_per_df.to_sql('Gross_rent_percentage',con=engine,if_exists='append',index_label='State')

value_est_df.to_sql(name='Value_estimate', con=engine,if_exists='append',index_label='State')

occupancy_per_df.to_sql(name='Occupancy_percentage', con=engine,if_exists='append',index_label='State')
```
